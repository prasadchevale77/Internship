### 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)Rank, B) Name, C) Artist, D) Upload date, E) Views 

```python
!pip install selenium
```

```python
import selenium
import pandas as pd
from selenium import webdriver

import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python

Name = []
Artist = []
Upload_date = []
Views = []
```

```python


names = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in names:
    Name.append(i.text)
    
    
    
artist_name = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for j in artist_name:
    Artist.append(j.text)
    
    
    
date = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for k in date:
    Upload_date.append(k.text)
    
    
    
view = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for l in view:
    Views.append(l.text)
    
```

```python
print(len(Name))
print(len(Artist))
print(len(Upload_date))
print(len(Views))
```

```python
df = pd.DataFrame({'Name': Name, 'Artist Name': Artist, 'Upload Date': Upload_date, 'Views': Views})
df
```

 B) Place
 C) Date
 D) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code. 

```python
!pip install selenium
```

```python
import selenium
import pandas as pd
from selenium import webdriver

import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
import time

from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.bcci.tv/")
```

```python
fixtures = driver.find_element(By.XPATH, '/html/body/header/div[3]/div[2]/ul/div[1]/a[2]')
fixtures.click()
```

```python
more_search = driver.find_element(By.XPATH, '/html/body/section/div/div/div/div/div/div[2]/div[2]/div[2]/div/button')
more_search.click()
```

```python
urls=[]

page_urls=driver.find_elements(By.XPATH, '//a[@class="match-btn match-card-btn ng-scope"]')
        
for url in page_urls:
    url = url.get_attribute('href')
    if url[0:4] == 'http':
        urls.append(url)
```

```python
len(urls)
```

```python
urls
```

```python
Series_name = []
Place_name = []
Date_ = []
Time_ = []
```

```python
for a in urls:
    driver.get(a)
    time.sleep(10)
    
    
    try:
        series_details = driver.find_element(By.XPATH, '//div[@class="tournament--name ng-binding ng-scope"][1]')
        Series_name.append(series_details.text)
    except NoSuchElementException as e:
        Series_name.append("--")   
        
        
    try:
        place_details = driver.find_element(By.XPATH, '//div[@class="matchVenue"]')
        Place_name.append(place_details.text)
    except NoSuchElementException as e:
        Place_name.append("--")
        
        
    
    try:
        date_details = driver.find_element(By.XPATH, '//div[@class="matchDate ng-binding ng-scope"]')
        Date_.append(date_details.text)
    except NoSuchElementException as e:
        Date_.append("--")
        
        
        
    
    try:
        time_details = driver.find_element(By.XPATH, '//div[@class="matchTm ng-binding"]')
        Time_.append(time_details.text)
    except NoSuchElementException as e:
        Time_.append("--")
```

```python
print(len(Series_name))
print(len(Place_name))
print(len(Date_))
print(len(Time_))
```

```python
df = pd.DataFrame({'Series name': Series_name, 'Place name': Place_name, 'Date': Date_, 'Time': Time_})
df
```

```python
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
```

```python
name_= driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/h3/a')
for i in name_ :
    Name.append(i.text)
    
    
year = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/h3/span[2]')
for j in year:
    Year_span.append(j.text)
    
    
genre_ = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/p/span[5]')
for k in genre_ :
    Genre.append(k.text)
    
    
time = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/p/span[3]')
for l in time:
    Run_time.append(l.text)
    
    
rating_ = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/div[1]/div[1]')
for m in rating_:
    Ratings.append(m.text)
    
    
    
vote = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/p[4]/span[2]')
for n in vote:
    Votes.append(n.text)
```

```python
print(len(Name))
print(len(Year_span))
print(len(Genre))
print(len(Run_time))
print(len(Ratings))
print(len(Votes))
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
dataset = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
dataset.click()
```

```python
Dataset_name =[]
Data_type =[]
Task =[]
Attribute_type =[]
No_of_instances =[]
No_of_attribute =[]
Year =[]
```

```python
expand_all = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[2]')
expand_all.click()
```

```python
Dataset name 
Data type 
Task 
Attribute type 
No of instances 
No of attribute 
Year
```

```python
name= driver.find_elements(By.XPATH, '//h2[@class="truncate text-primary"]/a')
for i in name :
    Dataset_name.append(i.text)
    
    
type_ = driver.find_elements(By.XPATH, '//div[@class="col-span-3 flex items-center gap-2"][2]/span')
for j in type_:
    Data_type.append(j.text)
    
    
task_ = driver.find_elements(By.XPATH, '//div[@class="col-span-3 flex items-center gap-2"][1]/span')
for k in task_ :
    Task.append(k.text)
    
    
attribute_type_ = driver.find_elements(By.XPATH, '//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[2]')
for l in attribute_type_:
    Attribute_type.append(l.text)
    

instance= driver.find_elements(By.XPATH, '//div[@class="col-span-3 flex items-center gap-2"][3]/span')
for m in instance:
    No_of_instances.append(m.text)
    
    
    
attribute = driver.find_elements(By.XPATH, '//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[1]')
for n in attribute:
    No_of_attribute.append(n.text)
    
    
    
year = driver.find_elements(By.XPATH, '//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[3]')
for a in year:
    Year.append(a.text)
```

```python
print(len(Dataset_name))
print(len(Data_type))
print(len(Task))
print(len(Attribute_type))
print(len(No_of_instances))
print(len(No_of_attribute))
print(len(Year))
```
