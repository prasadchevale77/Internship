### Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and salary filter.  
You have to scrape data for “Data Scientist” designation for first 10 job results.  
You have to scrape the job-title, job-location, company name, experience required.  
The location filter to be used is “Delhi/NCR”. The salary filter to be used is “3-6” lakhs  
The task will be done as shown in the below steps:  
1. first get the web page https://www.naukri.com/ 
2. Enter “Data Scientist” in “Skill, Designations, and Companies” field.  
3. Then click the search button.  
4. Then apply the location filter and salary filter by checking the respective boxes  
5. Then scrape the data for the first 10 jobs results you get.  
6. Finally create a dataframe of the scraped data.  
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
!pip install selenium
```

```python
import selenium
import pandas as pd
from selenium import webdriver

import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
## Opening the naukari page on automated chrome browser

driver.get("https://www.naukri.com/")
```

```python
## Entering designation and Location as required in the question-

designation = driver.find_element(By.CLASS_NAME, "suggestor-input")
designation.send_keys('Data Scientist')
```

```python
search = driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()
```

```python
#location filter

location = driver.find_element(By.XPATH, "/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[2]/div[2]/div[2]/label") 
location.click()
```

```python
# SALARY FILTER

salary = driver.find_element(By.XPATH, "/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[3]/div[2]/div[2]/label/p/span[1]")
salary.click()
```

```python
job_title = []
job_location = []
company_name = []
experience_required = []
```

```python
# scraping Job Title from the given page

title_tags = driver.find_elements(By.XPATH, '//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags:
    title=i.text
    job_title.append(title)
    
# scraping Job Location from the given page

location_tags = driver.find_elements(By.XPATH, '//span[@class="loc-wrap ver-line"]/span/span')
for i in location_tags:
    location=i.text
    job_location.append(location)


# scraping company name from the given page
company_tags = driver.find_elements(By.XPATH, '//div[@class=" row2"]/span/a[1]')
for i in company_tags:
    company = i.text
    company_name.append(company)

      
# Scraping Job Experience from the given page

experience_tags = driver.find_elements(By.XPATH, '//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-experience exp"]')
for i in experience_tags:
    experience = i.text
    experience_required.append(experience)
```

```python
print(len(job_title),len(job_location), len(company_name), len(experience_required))
```

```python
df = pd.DataFrame({'Title' : job_title[:10], 'Location' :job_location[:10], 'Company_name':company_name[:10], 'Experience' : experience_required[:10]})
df
```

### Q3: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link: 
https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART
As shown in the above page you have to scrape the tick marked attributes. These are: 
1. Rating 
2. Review summary 
3. Full review 
4. You have to scrape this data for first 100reviews. 
Note: All the steps required during scraping should be done through code only and not manually.

```python
!pip install selenium
```

```python
import selenium
import pandas as pd
from selenium import webdriver

import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
## Opening the naukari page on automated chrome browser

driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
```

```python
rating = []
Review = []
Full_review = []
```

```python
start = 0
end = 10

for page in range(start, end):

    rating_tags = driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq"]')
    for i in rating_tags:
        rating.append(i.text)
    

    review_tags = driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')
    for i in review_tags:
        Review.append(i.text)
    
    full_review_tags = driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]')
    for i in full_review_tags:
        Full_review.append(i.text)
        
   
    
    #to scrap data from next pages as well
    
    next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')  
                                                    
                                                
            
    next_button.click()
    time.sleep(5)
```

```python
print(len(rating),len(Review), len(Full_review))
```

```python
df = pd.DataFrame({'Rating': rating, 'Review': Review, 'Full Reviews': Full_review})
df
```

### Q6: Write a python program to scrape data for Top 1000 Quotes of All Time. 
The above task will be done in following steps: 
1. First get the webpagehttps://www.azquotes.com/ 
2. Click on Top Quote 
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
!pip install selenium
```

```python
import selenium
import pandas as pd
from selenium import webdriver

import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")
```

```python
top_quotes = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top_quotes.click()
```

```python
Quote = []
Author= []
Type_of_quotes = []
```

```python
start = 0
end = 10

for page in range( start, end):
    
    quote_tag = driver.find_elements(By.XPATH, '//div[@class="wrap-block"]/p/a[2]')
    for i in quote_tag:
        Quote.append(i.text)
        
    
    author_tag = driver.find_elements(By.XPATH, '//div[@class="author"]/a')
    for i in author_tag:
        Author.append(i.text)
        
    
    type_quotes = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in type_quotes:
        Type_of_quotes.append(i.text)
    
    
    
    next_button = driver.find_element(By.XPATH, '//li[@class="next"]/a')  
    next_button.click()
    

    time.sleep(3)
    
    
driver.quit()
```

```python
print(len(Quote), len(Author), len(Type_of_quotes))
```

```python
df = pd.DataFrame({'Quote': Quote, 'Author': Author, 'Type_of_quotes': Type_of_quotes})
df
```

### Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1  
scrap the mentioned data and make the DataFrame

```python
!pip install selenium
```

```python
import selenium
import pandas as pd
from selenium import webdriver

import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1")
```

```python
data = []
```

```python
table = driver.find_elements(By.XPATH, '//div[@class="TableData"]')

rows = table.find_elements(By.TAG_NAME, 'tr')

for row in rows:
    
    column = row.find_elemnts(By.TAG_NAME, 'td')
    
    if len(column) == 4:
        
        name = column[0].text
        born_dead = column[1].text
        term_of_office = column[2].text
        remarks = column[3].text
        data.append([name, born_dead, term_of_office, remarks])
```

```python
df = pd.DataFrame(data, columns=['Name', 'Born-Dead', 'Term of Office', 'Remarks'])
df
```
